```python
import numpy as np
import pandas as pd
import difflib
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df= pd.read_csv("movies.csv")
df.head()
```

```python
df.shape
```

```python
#feature selection for recommendations
selected_features= ["genres", "keywords","tagline","cast","director"]
```

```python
selected_features
```

```python
#replacing all null values with empty string 
for feature in selected_features:
     df[feature]=df[feature].fillna("")
```

```python
#combing all 5selected features

combined_features= df['genres']+df['keywords']+df['tagline']+df['cast']+df['director']
```

```python
combined_features
```

```python
#converting text data into feature vectors
vectorizer = TfidfVectorizer()
```

```python
feature_vector= vectorizer.fit_transform(combined_features)
print(feature_vector)
```

```python
#getting similarity score using cosine similarity 

similarity = cosine_similarity(feature_vector)
similarity
```

```python
similarity.shape
```

```python
#getting movie name from user
movie_name =input("Enter your fav movie")
```

```python
#creating list af all movies present in dataset
list_of_titles= df['title'].tolist()
```

```python
list_of_titles
```

```python
# finding the close match for movie name given by user 

find_close_match = difflib.get_close_matches(movie_name, list_of_titles)
```

```python
find_close_match
```

```python
close_match= find_close_match[0]
close_match
```

```python
#finding index of movie with title
index_of_movie= df[df.title==close_match]['index'].values[0]
index_of_movie
```

```python
#list of similar movie
similarity_score= list(enumerate(similarity[index_of_movie]))
similarity_score
```

```python
len(similarity_score)
```

```python
#sorting the movies based on similarity score in descending values
sorted_movies= sorted(similarity_score, key= lambda x:x[1], reverse= True)
```

```python
sorted_movies
```

```python
# print the name of similar movies based on index 
print("Movies suggested for you are  ")
i= 1
for movie in sorted_movies:
    index= movie[0]
    title_from_index = df[df.index==index]['title'].values[0]
    if(i<20):
        print(i, title_from_index)
        i+=1
```

```python
movie_name =input("Enter your fav movie")
list_of_titles= df['title'].tolist()
find_close_match = difflib.get_close_matches(movie_name, list_of_titles)
close_match= find_close_match[0]
index_of_movie= df[df.title==close_match]['index'].values[0]
similarity_score= list(enumerate(similarity[index_of_movie]))
sorted_movies= sorted(similarity_score, key= lambda x:x[1], reverse= True)

print("Movies suggested for you are  ")
i= 1
for movie in sorted_movies:
    index= movie[0]
    title_from_index = df[df.index==index]['title'].values[0]
    if(i<20):
        print(i, title_from_index)
        i+=1
```
